```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from keras import layers
import matplotlib.pyplot as plt
import time
```

```python
train_ds = keras.utils.image_dataset_from_directory(
    r'C:\Users\pedro.quinet\Desktop\IC-UFRJ-Zanini\IC-UFRJ-Zanini\transfer_learning_reference_2\treino',
    labels="inferred",
    label_mode="int",
    class_names=None,
    color_mode="rgb",
    batch_size=32,
    shuffle=True,
    validation_split=0.1,
    subset="training",
    seed = 123)

validation_ds = keras.utils.image_dataset_from_directory(
    r'C:\Users\pedro.quinet\Desktop\IC-UFRJ-Zanini\IC-UFRJ-Zanini\transfer_learning_reference_2\treino',
    labels="inferred",
    label_mode="int",
    class_names=None,
    color_mode="rgb",
    batch_size=32,
    shuffle=True,
    validation_split=0.1,
    subset="validation",
    seed = 123)
```

```python
# Em bachts de 32
print(f"Number of training samples: {train_ds.cardinality()}")
print(f"Number of validation samples: {validation_ds.cardinality()}")
```

```python
resize_fn = keras.layers.Resizing(150, 150)

train_ds = train_ds.map(lambda x, y: (resize_fn(x), y))
validation_ds = validation_ds.map(lambda x, y: (resize_fn(x), y))
```

```python
augmentation_layers = [
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.1),
]


def data_augmentation(x):
    for layer in augmentation_layers:
        x = layer(x)
    return x


train_ds = train_ds.map(lambda x, y: (data_augmentation(x), y))
```

```python
for images, labels in train_ds.take(1):
    plt.figure(figsize=(10, 10))
    first_image = images[0]
    for i in range(9):
        ax = plt.subplot(3, 3, i + 1)
        augmented_image = data_augmentation(np.expand_dims(first_image, 0))
        plt.imshow(np.array(augmented_image[0]).astype("int32"))
        plt.title(int(labels[0]))
        plt.axis("off")
```

```python
base_model = keras.applications.Xception(
    weights="imagenet",  # Carregando os pesos do modelo pr-treinado no dataset imaginet.
    input_shape=(150, 150, 3),
    include_top=False,  # Tirar o classificador final do modelo
)

# Congelando o modelo base
base_model.trainable = False

# O input do modelo
inputs = keras.Input(shape=(150, 150, 3))

# O Xception pr treinado requer que os insputs tenham essa escala.
# de (0, 255) para (-1., +1.).
scale_layer = keras.layers.Rescaling(scale=1 / 127.5, offset=-1)
x = scale_layer(inputs)

x = base_model(x, training=False) # Passar o training = False para deixar o batchnorming layer do base_model em modo inferncia.


x = keras.layers.GlobalAveragePooling2D()(x)
x = keras.layers.Dropout(0.2)(x)  # Regularizar com dropout

outputs = keras.layers.Dense(75, activation="softmax")(x) # sada com 75 classes pois temos 75 espcies de borboletas no dataset.

# Criar o modelo final
model = keras.Model(inputs, outputs)

model.compile(
    optimizer=keras.optimizers.Adam(),
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"],
)


model.summary(show_trainable=True)
```

```python
# Callback para pegar os tempos

class TimingCallback(tf.keras.callbacks.Callback):
    """Callback para registrar o tempo gasto em cada poca."""
    
    def __init__(self, logs={}):
        self.times = [] # Lista para armazenar os tempos de cada poca

    def on_epoch_begin(self, epoch, logs={}):
        """ chamado no incio de uma poca."""
        self.starttime = time.time() # Registra o tempo de incio

    def on_epoch_end(self, epoch, logs={}):
        """ chamado no final de uma poca."""
        # Calcula a durao e adiciona  lista
        duration = time.time() - self.starttime
        self.times.append(duration)

time_callback = TimingCallback()
```

```python
epochs = 10
print("Fitting the top layer of the model")
history = model.fit(train_ds, epochs=epochs, validation_data=validation_ds, callbacks=[time_callback])
```

```python
# Extrair os dados do objeto history
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']
epoch_times = time_callback.times
epochs_range = range(1, len(acc) + 1) # S para no comear do 0
```

```python
# No vale a pena plotar tudo junto

plt.figure(figsize=(12, 4))
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.plot(epochs_range, epoch_times, label='Epoch Time', linestyle='--')
plt.title('No  muito til botar todas as informaes juntas')
plt.show()
```

```python

# Criar a figura e os eixos para os plots
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(12, 18))
fig.suptitle('Anlise do Treinamento')

# Plot 1: Acurcia de Treino e Validao
ax1.plot(epochs_range, acc, 'o-', label='Acurcia de Treino')
ax1.plot(epochs_range, val_acc, 'o--', label='Acurcia de Validao')
ax1.set_title('Acurcia por epoch')
ax1.set_xlabel('epoch')
ax1.set_ylabel('Acurcia')
ax1.legend(loc='lower right') # Para no atrapalhar o grfico
ax1.set_xticks(epochs_range)


# Plot 2: Loss de Treino e Validao
ax2.plot(epochs_range, loss, 'o-', label='Perda de Treino')
ax2.plot(epochs_range, val_loss, 'o--', label='Perda de Validao')
ax2.set_title('Perda (Loss) por epoch')
ax2.set_xlabel('epoch')
ax2.set_ylabel('Perda')
ax2.legend(loc='upper right') # Para no atrapalhar o grfico
ax2.set_xticks(epochs_range)


# Plot 3: Tempo por poca
ax3.bar(epochs_range, epoch_times)
ax3.set_title('Tempo de Treinamento por epoch')
ax3.set_xlabel('epoch')
ax3.set_ylabel('Tempo (segundos)')
ax3.set_xticks(epochs_range)

# Ajustar o layout e exibir os plots
plt.tight_layout(rect=[0, 0.03, 1, 0.97])
plt.show()
```

```python
# Treinar o modelo completo 

time_callback = TimingCallback() # Poder pegar os tempos de treinamento
base_model.trainable = True
model.summary(show_trainable=True)

model.compile(
    optimizer=keras.optimizers.Adam(1e-5),  # Low learning rate
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"],
)

epochs = 5
print("Fitting the end-to-end model")
history = model.fit(train_ds, epochs=epochs, validation_data=validation_ds, callbacks=[time_callback])
```

```python
# Extrair os dados do objeto history
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']
epoch_times = time_callback.times
epochs_range = range(1, len(acc) + 1) # S para no comear do 0
```

```python
# Criar a figura e os eixos para os plots
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(12, 18))
fig.suptitle('Anlise do Treinamento (Modelo Completo)')

# Plot 1: Acurcia de Treino e Validao
ax1.plot(epochs_range, acc, 'o-', label='Acurcia de Treino')
ax1.plot(epochs_range, val_acc, 'o--', label='Acurcia de Validao')
ax1.set_title('Acurcia por epoch')
ax1.set_xlabel('epoch')
ax1.set_ylabel('Acurcia')
ax1.legend(loc='lower right') # Para no atrapalhar o grfico
ax1.set_xticks(epochs_range)


# Plot 2: Loss de Treino e Validao
ax2.plot(epochs_range, loss, 'o-', label='Perda de Treino')
ax2.plot(epochs_range, val_loss, 'o--', label='Perda de Validao')
ax2.set_title('Perda (Loss) por epoch')
ax2.set_xlabel('epoch')
ax2.set_ylabel('Perda')
ax2.legend(loc='upper right') # Para no atrapalhar o grfico
ax2.set_xticks(epochs_range)


# Plot 3: Tempo por poca
ax3.bar(epochs_range, epoch_times)
ax3.set_title('Tempo de Treinamento por epoch')
ax3.set_xlabel('epoch')
ax3.set_ylabel('Tempo (segundos)')
ax3.set_xticks(epochs_range)

# Ajustar o layout e exibir os plots
plt.tight_layout(rect=[0, 0.03, 1, 0.97])
plt.show()
```

```python
# Os dados de teste no tem labels :-()

print("Validation dataset evaluation")
model.evaluate(validation_ds) # -> Loss, accuracy
```
